```python
import pandas as pd
import numpy as np
from apyori import apriori
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import csv
```

```python
bank_df = pd.read_csv("data/Pre_Clean_credit_card_data.csv")
default1_df=bank_df.loc[bank_df['default payment in Jan 2022'] == 1]
#default_df=default1_df.drop(['default payment in Jan 2022'],axis=1)
default_df=default1_df.drop(['ID','default payment in Jan 2022','LIMIT_BAL','AGE','HOUSING','LOAN'],axis=1)
#print(default_df)

defaultedList = default_df.values.tolist()
#print(defaultedList)

te = TransactionEncoder()
te_ary = te.fit(defaultedList).transform(defaultedList)
final_df = pd.DataFrame(te_ary, columns=te.columns_)
print(final_df)


# Building the model
frq_items = apriori(final_df, min_support=0.3, use_colnames=True)
frq_items['length'] = frq_items['itemsets'].apply(lambda x: len(x))
print(frq_items)


# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending=[False, False])
print(rules.head())
```
